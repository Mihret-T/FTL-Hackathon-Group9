```python
import os
import cv2
import pytesseract
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from ultralytics import YOLO

# Define the path to your test images folder
test_images_folder = "counterfeit_med_detection-4/test/images/"

# Get the first image in the test folder
test_image = os.path.join(test_images_folder, os.listdir(test_images_folder)[0])

# Print to verify
print("Using test image:", test_image)

# Load the image
image = Image.open(test_image)

# Convert from PIL to OpenCV format (if needed)
image_cv2 = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

# Display the image using Matplotlib
plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.axis("off")  # Hide axis labels
plt.title("Test Image - Counterfeit Drug Detection")
plt.show()


# # Load the test image
# image = Image.open(test_image)
# image = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

# Apply OCR
text = pytesseract.image_to_string(image)

# Display extracted text
print("📌 Extracted Text from Image:")
print(text)

# Load trained model
model = YOLO("runs/detect/train9/weights/best.pt")  # Update if needed

# Run detection
results = model(test_image)

# Show detected image
results[0].show()

# Convert to image format and display using Matplotlib
plt.figure(figsize=(6, 6))
plt.imshow(results[0].plot())  # ✅ This will show the image with bounding boxes
plt.axis("off")
plt.title("YOLOv8 Detection - Counterfeit Drug Detection")
plt.show()
```
